# Filtros de diferencia

## 1. Prewitt

```
img = mpimg.imread('im0002.tif')[:,:,1]

#Crea el kernel
mask_prewitt_x = np.array([[-1 , 0 , 1],[-1 , 0 , 1],[-1 , 0 , 1]])
mask_prewitt_y = np.array([[-1 , -1 , -1],[0 , 0 , 0],[1 , 1 , 1]])

#Filtra la imagen utilizando el kernel anterior
dst_x = cv2.filter2D(img,-1,mask_prewitt_x)
dst_y = cv2.filter2D(img,-1,mask_prewitt_y)

plt.figure(1,figsize=(16,20))
plt.subplot(131),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(132),plt.axis('off'),plt.imshow(dst_x,cmap='gray'),plt.title('filtro de Prewitt X')
plt.subplot(133),plt.axis('off'),plt.imshow(dst_y,cmap='gray'),plt.title('filtro de Prewitt Y')

print('Kernel Prewitt X: \n')
print(mask_prewitt_x)
print('\n')
print('Kernel Prewitt Y: \n')
print(mask_prewitt_y)
```

```
# Este kernel resalta las diferencias entre pixeles de la imagen
```

```
# Filtro de Prewitt X: visualizacion en el eje X se coloca un foco 
# Filtro de Prewitt Y: visualizacion en el eje Y se coloca un foco 
```

## 2. Sobel

```
img = mpimg.imread('im0002.tif')[:,:,1]

#Crea el kernel
mask_sobel_x = np.array([[-1 , 0 , 1],[-2 , 0 , 2],[-1 , 0 , 1]])
mask_sobel_y = np.array([[-1 , -2 , -1],[0 , 0 , 0],[1 , 2 , 1]])

#Filtra la imagen utilizando el kernel anterior
dst_x = cv2.filter2D(img,-1,mask_sobel_x)
dst_y = cv2.filter2D(img,-1,mask_sobel_y)

plt.figure(1,figsize=(16,20))
plt.subplot(131),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(132),plt.axis('off'),plt.imshow(dst_x,cmap='gray'),plt.title('filtro de Sobel X')
plt.subplot(133),plt.axis('off'),plt.imshow(dst_y,cmap='gray'),plt.title('filtro de Sobel Y')

print('Kernel Sobel x: \n')
print(mask_sobel_x)
print('\n')
print('Kernel Sobel y: \n')
print(mask_sobel_y)
```

```
# A diferencia del Filtro de Prewwit le da mas importancia al centro , las venas se ven un poco mas gruesas
```

## 3. Roberts

```
# Los contornos se captan alponer un punto de iluminacion en las esquinas
img = mpimg.imread('im0002.tif')[:,:,1]

#Crea el kernel
mask_robert_1 = np.array([[0 , 1],[-1 , 0]])
mask_robert_2 = np.array([[1 , 0],[0 , -1]])

#Filtra la imagen utilizando el kernel anterior
dst_1 = cv2.filter2D(img,-1,mask_robert_1)
dst_2 = cv2.filter2D(img,-1,mask_robert_2)

plt.figure(1,figsize=(16,20))
plt.subplot(131),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(132),plt.axis('off'),plt.imshow(dst_1,cmap='gray'),plt.title('filtro de Robert 1')
plt.subplot(133),plt.axis('off'),plt.imshow(dst_2,cmap='gray'),plt.title('filtro de Robert 2')

print('Kernel Robert 1: \n')
print(mask_robert_1)
print('\n')
print('Kernel Robert 2: \n')
print(mask_robert_2)
```

## 4. Laplaciano

```
img = mpimg.imread('im0002.tif')[:,:,1]

#Crea el kernel
mask_laplaciano_1 = np.array([[0 , 1 , 0],[1 , -4 , 1],[0 , 1 , 0]])
mask_laplaciano_2 = np.array([[1 , 1 , 1],[1 , -8 , 1],[1 , 1 , 1]])

#Filtra la imagen utilizando el kernel anterior
dst_1 = cv2.filter2D(img,-1,mask_laplaciano_1)
dst_2 = cv2.filter2D(img,-1,mask_laplaciano_2)

plt.figure(1,figsize=(16,20))
plt.subplot(131),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(132),plt.axis('off'),plt.imshow(dst_1,cmap='gray'),plt.title('filtro de Laplaciano basico')
plt.subplot(133),plt.axis('off'),plt.imshow(dst_2,cmap='gray'),plt.title('filtro de Laplaciano con contribucion de las esquinas')

print('Kernel laplaciano 1: \n')
print(mask_laplaciano_1)
print('\n')
print('Kernel laplaciano 2: \n')
print(mask_laplaciano_2)
```

## 5. Canny

```
img = mpimg.imread('keyb.tif')

edge_img = cv2.Canny(img,100,300)

plt.figure(1,figsize=(16,20))
plt.subplot(121),plt.imshow(img,cmap='gray'),plt.title('Original')

plt.subplot(122),plt.imshow(edge_img,cmap='gray'),plt.title('filtro de Canny')
```

```
# Los bordes se detectan al aplicar el filtro laplaciano a una imagen de forma de en la "diagonal"
# se observan valores altos(255) y alrededor valores bajos(0)
```
