# Filtros No lineales

# 1. Maximo

```
img = mpimg.imread('keyb.tif')
img_fil_max = ndimage.maximum_filter(img, size=5)

plt.figure(1,figsize=(16,20))
plt.subplot(121),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(122),plt.axis('off'),plt.imshow(img_fil_max,cmap='gray'),plt.title('filtro no lineal - Maximo')
plt.show()
```

# 2. Minimo

```
img = mpimg.imread('keyb.tif')
img_fil_max = ndimage.minimum_filter(img, size=10)

plt.figure(1,figsize=(16,20))
plt.subplot(121),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(122),plt.axis('off'),plt.imshow(img_fil_max,cmap='gray'),plt.title('Filtro no lineal - Minimo')
plt.show()
```

# 3. Mediana

```
img = mpimg.imread('keyb.tif')
img_fil_max = ndimage.median_filter(img, size=10)

plt.figure(1,figsize=(16,20))
plt.subplot(121),plt.axis('off'),plt.imshow(img,cmap='gray'),plt.title('Original')
plt.subplot(122),plt.axis('off'),plt.imshow(img_fil_max,cmap='gray'),plt.title('filtro no lineal - Mediana')
plt.show()
```
