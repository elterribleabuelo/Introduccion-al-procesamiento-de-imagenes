## 1.2 2D

```
a  = [ [1.5, 2.3, 5.2] ,  [4.2, 5.6, 4.4]  ]
a1 = np.array( a )

b = ( (1,2) , (3,'hola') )
b1 = np.array( b )

print(a)
print(a1)
print('\n')
print(b)
print(b1)
```

```
a = [ [1.5, 2.3, 5.2] ,  [4.2, 5.6, 4.4]  ]
print(a)
print("\n")

a = np.array( a )
print(a)
print(np.shape(a))
print("\n")

print(a[:])
print("\n")

print(a[0][1])
print("\n")

print(a[-1])
print("\n")

print(a[0][0:2])
print("\n")
```

# 2. Herramientas de numpy

## 2.1 Creacion del array

```
a = np.array([2,3,4])
b = np.zeros((2, 3),dtype="uint8")
c1 = np.ones((2,5))
c2 = np.ones((1,5))
c3 = np.ones(5)

print(a,'\n')
print(b,'\n')
print(c1,'\n')
print(c2,'\n')
print(c3,'\n')
```

```
a1 = np.arange(10)
a2 = np.arange(2, 10, dtype=float)
a3 = np.arange(2, 3, 0.1)

print(a1,'\n')
print(a2,'\n')
print(a3,'\n')
```

```
a = np.linspace(1., 4., 6)
b = np.indices((4,2))

print(a,'\n')
print(b,'\n')
```

## 2.2 Caracteristicas del array

```
a = np.array([ [2 , 4] , [3 , 6] , [5 , 8]])
#a = np.array([ [2 , 4] ])
#a = np.array([2 , 4])

print('Dimension:', a.shape)
print('Número de filas:', a.shape[0])
print('Número de columnas:', a.shape[1])
print('Elementos, a=\n', a)
```

```
print(a.dtype)
print(a.ndim)
print(a.max())
print(a.min())
print(a.sum())
print(np.cumsum(a))

b = a//a.max()
print(b)
```

## 2.3 Operaciones y transformaciones en array

```
a = np.array([[1, 0], [0, 1]])
b = np.array([[4, 1], [2, 2]])
print(np.dot(a, b))
```

```
a = np.array([ [2 , 4] , [3 , 6] , [5 , 8]])
print(a)
print('\n')

b = a.reshape(2,3)
print(b)
```

```
a = np.arange(2,10)
print(a)
b=np.where(a < 7)
print(b)
```

```
a = np.arange(10)
print(a)
b=np.where(a < 5, a,a*2)
print(b)
```
