# 1. Reducion - Aumento

```
img = mpimg.imread('gato_ponido.jpg')

dim = img.shape
print(dim)

# Reducimos a la 64ava parte
img_res = cv2.resize( img , dsize=(int(dim[1]/8), int(dim[0]/8)), # que escala necesita (columnas, filas)
                     interpolation=cv2.INTER_CUBIC # metodo de interpolacion para crear datos
                     )
print(img_res.shape)

img_final=cv2.resize(img_res, dsize=(dim[1] ,dim[0] ), interpolation=cv2.INTER_CUBIC)
print(img_final.shape)

plt.figure(1,figsize=(16,12))

plt.subplot(1, 3, 1)
plt.title(str(img.shape[0])+'x'+str(img.shape[1]))
plt.imshow(img,cmap='gray')

plt.subplot(1, 3, 2)
plt.title(str(img_res.shape[0])+'x'+str(img_res.shape[1]))
plt.imshow(img_res,cmap='gray')

plt.subplot(1, 3, 3)
plt.title(str(img_final.shape[0])+'x'+str(img_final.shape[1]))
plt.imshow(img_final,cmap='gray')
```

# 2. Aumento - Reduccion

```
img = mpimg.imread('gato_ponido.jpg')

dim = img.shape
print(dim)

img_res = cv2.resize( img , dsize=(int(dim[1]*8), int(dim[0]*8)), interpolation=cv2.INTER_CUBIC)
print(img_res.shape)

img_final=cv2.resize(img_res, dsize=(dim[1] ,dim[0] ), interpolation=cv2.INTER_CUBIC)
print(img_final.shape)


plt.figure(1,figsize=(16,12))

plt.subplot(1, 3, 1)
plt.title(str(img.shape[0])+'x'+str(img.shape[1]))
plt.imshow(img,cmap='gray')

plt.subplot(1, 3, 2)
plt.title(str(img_res.shape[0])+'x'+str(img_res.shape[1]))
plt.imshow(img_res,cmap='gray')

plt.subplot(1, 3, 3)
plt.title(str(img_final.shape[0])+'x'+str(img_final.shape[1]))
plt.imshow(img_final,cmap='gray')
```
