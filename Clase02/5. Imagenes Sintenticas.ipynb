# 2. Normalizar

La normalizacion consiste en convertir los datos para que varien entre 0 y 255. 

\begin{equation}
{x_{norm} ={\frac {(x-min(x))*255}{max(x) - min(x)}}}
\end{equation}

```
faux = f.astype(float)
print(faux)
minimum = faux.min()
maximum = faux.max()

g = (faux-minimum)*(255) / (maximum-minimum)
g = g.reshape(f.shape).astype(np.uint8)
plt.imshow(g,cmap='gray')
#plt.imshow(f,cmap='gray')
```

# 3. Implementar la funcion para normalizar (clase 5 minutos) 1 punto

```
def Normalizar(f):
  faux = np.ravel(f).astype(float)
  minimum = faux.min()
  maximum = faux.max()
  g = (faux-minimum)*(255) / (maximum-minimum)
  r = g.reshape(f.shape).astype(np.uint8)
  return(r)
```

## 4.1 Ejemplos 1

```
### IMAGEN 1 ###

r,c = np.indices( (600, 600) ) # con (100,100) no se aprecian todas las tonaliddaes
f=r
print(f)
f_normalizado=Normalizar(f)  
print(f_normalizado)
plt.imshow(f_normalizado,cmap='gray')
```

```
### IMAGEN 3 ###

r,c = np.indices( (300, 300) )
f = ((r**2 + c**2) > 150**2) # Ecuacion de la circunferencia
print(r) # indices i
print(c) # indices j
f_normalizado=Normalizar(f)  
#print(f_normalizado)
plt.imshow(f_normalizado,cmap='gray')
```

## 4.5 Ejemplos 3

```
### IMAGEN 1 ###

# Paraboloide circular
control=10
# Para tener mayor detalle aumentar de 20 a 200
rows = np.linspace(control*-1, control, 200) 
cols = np.linspace(-1*control, control, 200) 
r, c = np.meshgrid(rows, cols, indexing='ij')
#print(rows)
#print("hola")
#print(r)
#print(c)

z=r**2+c**2
# Debe estar normalizado para realizar operaciones entre imagenes
f_normalizado=Normalizar(z)
plt.imshow(f_normalizado,cmap='gray')
```

```
### IMAGEN 2 ###
# Paraboloide eliptico
control=10
# 400-> resolucion
# Creamos los indices
rows = np.linspace(control*-1, control, 400) # coordenadas de lineas
cols = np.linspace(-1*control, control, 400) # coordenadas de columnas
r, c = np.meshgrid(rows, cols, indexing='ij')
z=(r**2)/2+(c**2)/4
f_normalizado=Normalizar(z)  
plt.imshow(f_normalizado,cmap='gray')
```

```
### IMAGEN 3 ###

control=10
rows = np.linspace(control*-1, control, 400) # coordenadas das linhas
cols = np.linspace(-1*control, control, 400) # coordenadas das colunas
r, c = np.meshgrid(rows, cols, indexing='ij') # creamos los indices de una cuadricula de 400x400 centrado en 0
e = np.spacing(1) # Valor aleatorio minusculo
print(e)
z = np.sin(r**2 + c**2 + e) / (r**2 + c**2 + e)  ## sen(x)/x-----sen(x+e)/(x+e)
f_normalizado=Normalizar(z)  
#print(f_normalizado)
plt.imshow(f_normalizado,cmap='gray')
```

```
### IMAGEN 4 ###

control=10
rows = np.linspace(control*-1, control, 400) # coordenadas das linhas
cols = np.linspace(-1*control, control, 400) # coordenadas das colunas
r, c = np.meshgrid(rows, cols, indexing='ij')
e = np.spacing(1)
print(e)
z = np.sin((r**2)/2 + (c**2)/4 + e) / ((r**2)/2 + (c**2)/4 + e)
f_normalizado=Normalizar(z)  
#print(f_normalizado)
plt.imshow(f_normalizado,cmap='gray')
```
