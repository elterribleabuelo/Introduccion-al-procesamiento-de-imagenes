## 1.2 Manejo de la imagen como matriz.

### 1.2.1 Analisis de la imagen

```
a = mpimg.imread('machupicchu.jpg')
b = mpimg.imread('machupicchu.jpg')[0:10,0:10,0]

print(a.shape)
print("Imagen: \n")
print(a)
print('\n')

print(b.shape)
print('Imagen de entrada recortada: \n')
print(b)
print('\n')

# Maximo valor de toda la matriz
print('Maximo valor: ')
print(a.max())
print('\n')

# Promedio de  valores de toda la matriz
print('valor promedio: ')
print(a.mean())
print('\n')

# Suma de valores de toda la matriz
print('suma de valor: ')
print(a.sum())
print('\n')

print('Dimesiones de la imagen: ')
print(a.shape)
```

```
plt.figure(1,figsize=(8,10))
plt.subplot(1, 2, 1)
plt.imshow(a,cmap='gray')
plt.title('Imagen de entrada')

plt.subplot(1, 2, 2)
plt.imshow(b,cmap='gray')
plt.title('10 x 10')

print('Matriz b = ')
print(b)
```

```
print('Matriz b = ')
print(b)

# Maximo de las columnas
print('\n')
print(b.max(0))

# Maximo de las filas
print('\n')
print(b.max(1))

# Promedio de las columnas
print('\n')
print(b.mean(0))

# Minimo de las columnas
print('\n')
print(b.min(0))

# Suma de las columnas
print('\n')
print(b.sum(0))
```

```
# Dividir en 6 partes una imagen
def f1(imagen):
  tarea = mpimg.imread(imagen)
  Dim=np.shape(tarea)

  tareaG = tarea[:,:,1]

  a_mitad_1=tarea[:,0:int(Dim[0]/2),1]
  Dim1 = np.shape(a_mitad_1)

  a_mitad_2=tarea[:,int(Dim[0]/2):,1]
  Dim2 = np.shape(a_mitad_2)

  plt.figure(1,figsize=(8,6))

  a_parte_1 = a_mitad_1[0:int(Dim1[0]/3),:]
  #plt.subplot(3, 2, 1)
  #plt.imshow(a_parte_1,cmap = 'gray')

  a_parte_2 = a_mitad_1[int(Dim1[0]/3):int(2*Dim1[0]/3),:]
  #plt.subplot(3, 2, 3)
  #plt.imshow(a_parte_2,cmap = 'gray')

  a_parte_3 = a_mitad_1[int(2*Dim1[0]/3):,:]
  #plt.subplot(3, 2, 5)
  #plt.imshow(a_parte_3,cmap = 'gray')

  a_parte_4 = a_mitad_2[0:int(Dim2[0]/3),:]
  #plt.subplot(3, 2, 2)
  #plt.imshow(a_parte_4,cmap = 'gray')

  a_parte_5 = a_mitad_2[int(Dim2[0]/3):int(2*Dim2[0]/3),:]
  #plt.subplot(3, 2, 4)
  #plt.imshow(a_parte_5,cmap = 'gray')

  a_parte_6 = a_mitad_2[int(2*Dim2[0]/3):,:]
  #plt.subplot(3, 2, 6)
  #plt.imshow(a_parte_6,cmap = 'gray')
  
  """eje_y1 = np.concatenate((a_parte_1,
                           a_parte_3,
                           a_parte_2),
                          axis = 0)
  eje_y2 = np.concatenate((a_parte_4,
                           a_parte_6,
                           a_parte_5),
                          axis = 0)
  resultado = np.concatenate((eje_y2,
                              eje_y1),
                             axis = 1)"""
  #plt.imshow(resultado,cmap ='gray')
  ## Imagen contenedor
  N = Dim[0]
  M = Dim[1]
  j1 = np.zeros((N, M), np.int64)
  # Ubicamos las partes
  j1[:a_parte_4.shape[0],:a_parte_4.shape[1]] = a_parte_4
  j1[:a_parte_4.shape[0],a_parte_4.shape[1]:a_parte_4.shape[1] + a_parte_1.shape[1]] = a_parte_1

  j1[a_parte_4.shape[0]:a_parte_4.shape[0] + a_parte_2.shape[0], :a_parte_2.shape[1] ] = a_parte_2
  j1[a_parte_4.shape[0]:a_parte_4.shape[0] + a_parte_2.shape[0], a_parte_2.shape[1]:a_parte_2.shape[1] + a_parte_5.shape[1]] = a_parte_5

  j1[a_parte_4.shape[0] + a_parte_2.shape[0]:a_parte_4.shape[0] + a_parte_2.shape[0] + a_parte_6.shape[0],:a_parte_6.shape[1]] = a_parte_6
  j1[a_parte_4.shape[0] + a_parte_2.shape[0]:a_parte_4.shape[0] + a_parte_2.shape[0] + a_parte_6.shape[0], a_parte_6.shape[1]:a_parte_6.shape[1] + a_parte_3.shape[1] ] = a_parte_3

  plt.subplot(1, 2, 1)
  plt.imshow(tareaG,cmap = 'gray')
  
  plt.subplot(1, 2, 2)
  plt.imshow(j1,cmap = 'gray')
```
