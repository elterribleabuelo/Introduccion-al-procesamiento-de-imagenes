```
import numpy as np
import matplotlib.pyplot as plt

n_datos=100
X = np.linspace(0, 2, n_datos)
y = 3 +2 * X + np.random.randn(n_datos)* 0.33

plt.figure(1,figsize=(21,6))
plt.subplot(1, 3, 1)
plt.title('Datos random con comportamiento Y=2X+3')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X,y,'b-')

plt.subplot(1, 3, 2)
plt.title('Datos random con comportamiento Y=2X+3')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X,y,'r.')

plt.subplot(1, 3, 3)
plt.title('Datos random con comportamiento Y=2X+3')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X,y,'g--')
plt.show()
```

```
n_datos=100
X1 = np.linspace(0, 2, n_datos)
y1 = 3 + 2 * X1 + np.random.randn(n_datos)* 0.33

X2 = np.linspace(0, 2, n_datos)
y2 = 3 - 2 * X2 + np.random.randn(n_datos)* 0.33

plt.figure(1,figsize=(18,6))
plt.subplot(1, 3, 1)
plt.title('Datos random con comportamiento Y=2X+3')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(X1,y1,'b-')
plt.plot(X2,y2,'r*')
```

```
# Fractales 

import numpy as np
import numba
import matplotlib.pyplot as plt

def py_julia_fractal(z_re, z_im, j,expo):
    '''Crea el grafico del fractal de Julia.'''
    for m in range(len(z_re)):
        for n in range(len(z_im)):
            z = z_re[m] + 1j * z_im[n]
            for t in range(256):
                z = z ** expo - 0.05 + 0.68j
                if np.abs(z) > 2.0:
                    j[m, n] = t
                    break
                    
jit_julia_fractal = numba.jit(nopython=True)(py_julia_fractal)

N = 1024
j1 = np.zeros((N, N), np.int64)
j2 = np.zeros((N, N), np.int64)
j3 = np.zeros((N, N), np.int64)
j4 = np.zeros((N, N), np.int64)

z_real = np.linspace(-1.5, 1.5, N)
z_imag = np.linspace(-1.5, 1.5, N)

jit_julia_fractal(z_real, z_imag, j1,2)
jit_julia_fractal(z_real, z_imag, j2,3)
jit_julia_fractal(z_real, z_imag, j3,4)
jit_julia_fractal(z_real, z_imag, j4,5)

plt.figure(1,figsize=(24, 6))

plt.subplot(1,4,1)
plt.imshow(j1, cmap=plt.cm.RdBu_r, extent=[-1.5, 1.5, -1.5, 1.5])
plt.xlabel("$\mathrm{Re}(z)$", fontsize=18)
plt.ylabel("$\mathrm{Im}(z)$", fontsize=18)


plt.subplot(1,4,2)
plt.imshow(j2, cmap=plt.cm.RdBu_r, extent=[-1.5, 1.5, -1.5, 1.5])
plt.xlabel("$\mathrm{Re}(z)$", fontsize=18)
plt.ylabel("$\mathrm{Im}(z)$", fontsize=18)


plt.subplot(1,4,3)
plt.imshow(j3, cmap=plt.cm.RdBu_r, extent=[-1.5, 1.5, -1.5, 1.5])
plt.xlabel("$\mathrm{Re}(z)$", fontsize=18)
plt.ylabel("$\mathrm{Im}(z)$", fontsize=18)

plt.subplot(1,4,4)
plt.imshow(j4, cmap=plt.cm.RdBu_r, extent=[-1.5, 1.5, -1.5, 1.5])
plt.xlabel("$\mathrm{Re}(z)$", fontsize=18)
plt.ylabel("$\mathrm{Im}(z)$", fontsize=18)
```
